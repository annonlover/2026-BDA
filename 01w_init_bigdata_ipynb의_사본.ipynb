```python
import pandas as pd
pop = pd.read_csv("popmonth.csv", encoding='cp949', index_col=0, thousands=',')
pop
```

```python
pop.columns = ['총 인구수', '새대 수', '세대당인구', '남자 인구수','여자 인구수', ' 남녀 비율']
pop.head()
```

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(pop['총 인구수'])
plt.xticks(rotation= 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pop1 = pop.sort_values('총 인구수')
sns.barplot(x=pop1.index, y='총 인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```
